```python
{
 "cells": [
  {
   "cell_type": "code",
   "execution_count": null,
   "metadata": {},
   "outputs": [],
   "source": [
    "import numpy as np\n",
    "import matplotlib.pyplot as plt\n",
    "\n",
    "# 物性値（温度 22.5℃）\n",
    "beta = 3.0e-4\n",
    "nu   = 0.9e-6\n",
    "alpha = 0.14e-6\n",
    "k = 0.6\n",
    "g = 9.81\n",
    "\n",
    "def calc_Ra(H, dT):\n",
    "    return g * beta * dT * H**3 / (nu * alpha)\n",
    "\n",
    "def calc_Nu(Ra, C):\n",
    "    return C * Ra**(1/3)\n",
    "\n",
    "def calc_U(Nu, H):\n",
    "    return Nu * k / H\n",
    "\n",
    "def visualize_field(mode):\n",
    "    nx, ny = 40, 40\n",
    "    x = np.linspace(0, 1, nx)\n",
    "    y = np.linspace(0, 1, ny)\n",
    "    X, Y = np.meshgrid(x, y)\n",
    "\n",
    "    plt.figure(figsize=(6,5))\n",
    "\n",
    "    if mode == \"Bottom heating\":\n",
    "        A = 1.0\n",
    "        u =  A * np.sin(np.pi * X) * np.cos(np.pi * Y)\n",
    "        v = -A * np.cos(np.pi * X) * np.sin(np.pi * Y)\n",
    "        speed = np.sqrt(u**2 + v**2)\n",
    "        plt.streamplot(X, Y, u, v, color=speed, cmap='turbo')\n",
    "        plt.title(\"Convection streamlines (Bottom heating)\")\n",
    "        plt.colorbar(label=\"Velocity magnitude\")\n",
    "    else:\n",
    "        T = 1 - Y\n",
    "        plt.contourf(X, Y, T, levels=20, cmap='coolwarm')\n",
    "        plt.title(\"Isotherms (Top heating)\")\n",
    "        plt.colorbar(label=\"Temperature\")\n",
    "\n",
    "    plt.gca().invert_yaxis()\n",
    "    plt.xlabel(\"x\")\n",
    "    plt.ylabel(\"y\")\n",
    "    plt.show()\n",
    "\n",
    "def run_model(H_mm=90, dT=15, C=0.069, mode=\"Bottom heating\"):\n",
    "    H = H_mm / 1000\n",
    "    if mode == \"Bottom heating\":\n",
    "        Ra = calc_Ra(H, dT)\n",
    "        Nu = calc_Nu(Ra, C)\n",
    "    else:\n",
    "        Ra = calc_Ra(H, dT)\n",
    "        Nu = 1.0\n",
    "\n",
    "    U = calc_U(Nu, H)\n",
    "    print(f\"Mode = {mode}\")\n",
    "    print(f\"Rayleigh number Ra = {Ra:.3e}\")\n",
    "    print(f\"Nusselt number Nu = {Nu:.2f}\")\n",
    "    print(f\"U-value = {U:.2f} W/(m2·K)\")\n",
    "\n",
    "    H_list = np.linspace(0.03, 0.12, 50)\n",
    "    U_list = []\n",
    "    for H_i in H_list:\n",
    "        Ra_i = calc_Ra(H_i, dT)\n",
    "        Nu_i = calc_Nu(Ra_i, C) if mode == \"Bottom heating\" else 1.0\n",
    "        U_list.append(calc_U(Nu_i, H_i))\n",
    "\n",
    "    plt.figure(figsize=(7,5))\n",
    "    plt.plot(H_list*1000, U_list, lw=2)\n",
    "    plt.axvline(H_mm, color='red', linestyle='--')\n",
    "    plt.xlabel(\"Thickness H [mm]\")\n",
    "    plt.ylabel(\"U-value [W/(m2·K)]\")\n",
    "    plt.title(\"Effect of water layer thickness on U-value\")\n",
    "    plt.grid(True)\n",
    "    plt.show()\n",
    "\n",
    "    visualize_field(mode)\n",
    "\n",
    "run_model(H_mm=90, dT=15, C=0.069, mode=\"Bottom heating\")\n"
   ]
  }
 ],
 "metadata": {
  "kernelspec": {
   "display_name": "Python 3",
   "language": "python",
   "name": "python3"
  },
  "language_info": {
   "name": "python",
   "version": "3.10"
  }
 },
 "nbformat": 4,
 "nbformat_minor": 5
}
```
