下の枠にマウスを移動すると表示される、直下の枠左上の三角印をクリックしてください。動作してグラフが表示される場合まで15秒ほどかかります。

```python
import numpy as np
import matplotlib.pyplot as plt

# -----------------------------------
# Physical properties (at 22.5°C)
# -----------------------------------
beta = 3.0e-4
nu   = 0.9e-6
alpha = 0.14e-6
k = 0.6
g = 9.81

# -----------------------------------
# Rayleigh number
# -----------------------------------
def calc_rayleigh(H, dT):
    return g * beta * dT * H**3 / (nu * alpha)

# -----------------------------------
# Nusselt number
# -----------------------------------
def calc_nusselt(Ra, C):
    return C * Ra**(1/3)

# -----------------------------------
# Heat transfer coefficient U
# -----------------------------------
def calc_uvalue(Nu, H):
    return Nu * k / H

# -----------------------------------
# Flow field visualization
# -----------------------------------
def visualize_field(mode):
    nx, ny = 40, 40
    x = np.linspace(0, 1, nx)
    y = np.linspace(0, 1, ny)
    X, Y = np.meshgrid(x, y)

    plt.figure(figsize=(6,5))

    if mode == "Bottom heating":
        A = 1.0
        u =  A * np.sin(np.pi * X) * np.cos(np.pi * Y)
        v = -A * np.cos(np.pi * X) * np.sin(np.pi * Y)
        speed = np.sqrt(u**2 + v**2)

        plt.streamplot(X, Y, u, v, color=speed, cmap='turbo')
        plt.title("Convection Streamlines (Bottom Heating)")
        plt.colorbar(label="Velocity Magnitude")

    else:
        T = 1 - Y
        plt.contourf(X, Y, T, levels=20, cmap='coolwarm')
        plt.title("Isotherms (Top Heating)")
        plt.colorbar(label="Temperature")

    plt.gca().invert_yaxis()
    plt.xlabel("x")
    plt.ylabel("y")
    plt.show()

# -----------------------------------
# Main model
# -----------------------------------
def run_model(H_mm=90, dT=15, C=0.069, mode="Bottom heating"):
    H = H_mm / 1000

    if mode == "Bottom heating":
        Ra = calc_rayleigh(H, dT)
        Nu = calc_nusselt(Ra, C)
    else:
        Ra = calc_rayleigh(H, dT)
        Nu = 1.0

    U = calc_uvalue(Nu, H)

    print(f"Mode = {mode}")
    print(f"Rayleigh number = {Ra:.3e}")
    print(f"Nusselt number = {Nu:.2f}")
    print(f"U-value = {U:.2f} W/(m2·K)")

    # U-value vs thickness
    H_list = np.linspace(0.03, 0.12, 50)
    U_list = []

    for H_i in H_list:
        Ra_i = calc_rayleigh(H_i, dT)
        Nu_i = calc_nusselt(Ra_i, C) if mode == "Bottom heating" else 1.0
        U_list.append(calc_uvalue(Nu_i, H_i))

    plt.figure(figsize=(7,5))
    plt.plot(H_list*1000, U_list, lw=2)
    plt.axvline(H_mm, color='red', linestyle='--')
    plt.xlabel("Thickness H [mm]")
    plt.ylabel("U-value [W/(m2·K)]")
    plt.title("Effect of Water Layer Thickness on U-value")
    plt.grid(True)
    plt.show()

    visualize_field(mode)

# -----------------------------------
# Example execution
# -----------------------------------
run_model(H_mm=90, dT=15, C=0.069, mode="Bottom heating")
```
